```python
retail_data['Discount']=retail_data["TotalSales"]*0.1
print(retail_data['Discount'])     # adding additional discount column with 10%
print(retail_data)
```

```python
retail_data['Tax']=retail_data["TotalSales"]*0.15

print(retail_data)  # adding additional tax column with 15%
```
